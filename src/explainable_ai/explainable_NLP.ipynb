```python
import transformers
import shap
import torch
import datasets

import numpy as np
import scipy as sp

from transformers import AutoTokenizer, AutoModelForCausalLM
```

# Text Generation 1 - Bloomz 1.1B

```python
tokenizer = AutoTokenizer.from_pretrained("bigscience/bloomz-1b1")
model = AutoModelForCausalLM.from_pretrained("bigscience/bloomz-1b1")
```

```python
# set model decoder to true
model.config.is_decoder=True

text_gen_config = {
    "do_sample": True,
    "max_length": 20,
    "temperature": 0.7,
    "top_k": 50,
    "no_repeat_ngram_size": 2,
    "max_new_tokens": 25
}
# set text-generation params under task_specific_params
model.config.task_specific_params = dict()
model.config.task_specific_params["text-generation"] = text_gen_config
```

```python
s = ['Once upon a time, Columbus discovered India. ']
```

```python
shap_model = shap.models.TeacherForcing(model, tokenizer)
masker = shap.maskers.Text(tokenizer, mask_token="...", collapse_mask_token=True)
explainer = shap.Explainer(shap_model, tokenizer)
shap_values = explainer(s)
```

```python
shap.plots.text(shap_values)
```
